```python
import pandas as pd

ranked_component = pd.read_parquet('ranked_component.parquet')
ranked_component
```

```python
import matplotlib.pyplot as plt
from matplotlib import colors
divnorm=colors.TwoSlopeNorm(vmin=-0.7, vcenter=0.0, vmax=0.7)

weights_matrix = ranked_component[[mi for mi in ranked_component.columns if mi[1] == 'weight']]
plt.imshow(weights_matrix, cmap='seismic', norm=divnorm)
```

```python
i=3
df = ranked_component.iloc[:,2*i:(2*i+2)].copy().reset_index(drop=True)
df
```

```python
sorted_column_pair_dfs = []
for i in range(int(len(ranked_component.columns)/2)):
    df = ranked_component.iloc[:,2*i:(2*i+2)].copy().copy().reset_index(drop=True)
    df = df.sort_values(df.columns[0]).reset_index(drop=True)
    sorted_column_pair_dfs.append(df)

sorted_df = pd.concat(sorted_column_pair_dfs, axis=1)
```

```python

plt.rcParams['figure.figsize'] = [50,50]

sorted_weights_matrix = sorted_df[[mi for mi in sorted_df.columns if mi[1] == 'weight']].to_numpy()
plt.imshow(sorted_weights_matrix[:, 0:200], cmap='seismic', norm=divnorm)
```
